# My First Jupyter Notebook

In QSO370 we'll be using Jupyter Notebooks which embed both text and python code, to construct reproducible analyses. This is our first Jupyter Notebook for the course. Double-click into this text to see the cell that produces the text on the left and the formatted text on the right. Then follow along to see how formatting in Jupyter works.

We can type text freely and Jupyter will automatically format it for us in a presentable manner. 
Notice that we need to "skip a line" if we want to create a new paragraph.

We can also use some simple markdown code to format the text portion of our notebooks. You can surround text by a single asterisk (\*) to make it appear in *italics* or with two asterisks to make it appear in **bold** -- you'll notice that I needed to use a backslash prior to the asterisk symbol in order for it to actually appear in the text -- this is because the asterisk is a *special character* which needs to be *escaped* with the backslash if you want it to be printed into the notebook. You've already seen the use of the hastag (\#) to create a *level-one* header, or *title*). You can create level-two, -three, or higher-order headers just by using additional hastags. For example, preceed a line of text with two hashtags to create a level-two heading (section).

You may also want to create lists in your notebooks. If order is not important for my lists, we can create an unordred list using the + symbol to begin each list item line.
+ This is an item in my list
+ This is another item in my list

If order is important for the items in my list, we can number the items.
1. This is the first item in my list.
2. This is the second item in my list.
  + Hit tab and use a + sign to generate a sub-item in the list.

We can render the text cell into the notebook by using the key combination `Shift+Enter`. Notice how I formatted that? Using these single back-tics is a really convenient way to format unexecuted code or computer instructions within a text section.
